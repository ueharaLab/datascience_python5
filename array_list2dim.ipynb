### 3. 2次元配列と2重のfor文
以下2重のfor文になっている。内側のfor文はインデントしていることに注意  
#### **for文のネスト（入れ子）と呼ばれる**

```python
names_age = [ ['Bob',19], ['Ken',21], ['Alice',20] ]
for na in names_age:
    print(na)
    for k in na:
        print(k)
```
